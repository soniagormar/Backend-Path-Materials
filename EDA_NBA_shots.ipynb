**Exploratory Data Analisys 'NBA Shots'**

Analisys was carried on  dataset that was downloaded from [kaggle.com](https://www.kaggle.com/datasets/mexwell/nba-shots?resource=download).

Dataset includes infromation about shots taken during all of NBA games since 2023 till 2024.
This analisys focuses on 2023/2024 season.

\\
**Data Dictionary**


*   SEASON_1, SEASON_2 - indicates season
*   TEAM_ID, TEAM_NAME - unique identifier and name of the team
*   PLAYER_ID, PLAYER_NAME - unique identifier and name of player
*   POSITION_GROUP, POSITION - group and specific position of the player
*   GAME_DATE, GAME_ID - date and unique identifier of the game
*   HOME_TEAM - code for the home team
*   AWAY_TEAM - code for the away team
*   EVENT_TYPE - outcome of the shot (Missed Shot/MadeShot)
*   SHOT_MADE - boolean indicating shot success
*   ACTION_TYPE - type of shot attempted (layup, dunk, jump shot, etc.)
*   SHOT_TYPE - potential points from the shot(2PT or 3PT)
*   BASIC_ZONE - general area of the court where the shot was taken(Restricted Area, In the Paint (non-RA), Midrange, Left Corner 3, Right Corner 3, Above the Break, Backcourt)
*   ZONE_NAME - specific court area (left, left side center, center, right side center, right)
*   ZONE_ABB - abbreviation of the court area((L), (LC), (C), (RC), (R))
*   ZONE_RANGE - range fro which the shot was taken (Less than 8 ft., 8-16 ft. 16-24 ft. 24+ ft.)
*   LOC_X, LOC_Y - exact coordinates of the shot location in the court
*   SHOT_DISTANCE - distance of the shot from the basket (in feet)
*   QUARTER, MINS_LEFT, SECS_LEFT - shot timing infromation, specyfing quarter and time remaining

```python
players_shots = df.groupby(['PLAYER_NAME', 'ACTION_TYPE']).size().unstack(fill_value=0)
players_shots['FAVOURED_SHOT_TECHNIQUE'] = players_shots.idxmax(axis=1)
players_shots.reset_index()
```

```python
# ATTENTION! Data does not include information about players transfers between teams

players_accuracy = df.groupby(['PLAYER_NAME', 'EVENT_TYPE']).size().unstack(fill_value=0)
players_accuracy['SHOTS_TAKEN'] = players_accuracy['Made Shot'] + players_accuracy['Missed Shot']
players_accuracy['ACCURACY %'] = ((players_accuracy['Made Shot'] / (players_accuracy['SHOTS_TAKEN']))*100).round(2)

players_accuracy = players_accuracy.sort_values(by='PLAYER_NAME')

players_accuracy = players_accuracy.merge(players_shots[['FAVOURED_SHOT_TECHNIQUE']], on='PLAYER_NAME', how='left')

fav_position = df.groupby(['PLAYER_NAME', 'POSITION']).size().unstack().idxmax(axis=1).to_frame(name='FAVOURED_POSITION')

players_accuracy = players_accuracy.merge(fav_position, on = 'PLAYER_NAME', how = 'left')

players_accuracy.info()
```

```python
fig = px.scatter(players_accuracy,
                 x = 'SHOTS_TAKEN',
                 y = 'ACCURACY %',
                 title = 'Distribution of players\' accuracy based on the number of shots attempted.',
                 labels = {'ACCURACY %' : 'Accuracy', 'SHOTS_TAKEN' : 'Number of shots taken', 'FAVOURED_POSITION' : 'Players\' position'},
                 color = 'FAVOURED_POSITION'
                 )

mean_accuracy = players_accuracy['ACCURACY %'].mean()

fig.add_hline(
    y= mean_accuracy,
    line_dash="dash",
    line_color="red",
    annotation_text= f"Mean accuracy {mean_accuracy:.2f}",
    annotation_position='bottom right'
    )

fig.show()
```

```python
made_shots = df[df['SHOT_MADE'] == 1]
players_points = made_shots.groupby('PLAYER_NAME')['SHOT_VALUE'].sum().reset_index()
players_points = players_points.rename(columns={'SHOT_VALUE': 'TOTAL_POINTS'})
players_accuracy = players_accuracy.merge(players_points[['PLAYER_NAME', 'TOTAL_POINTS']], on='PLAYER_NAME', how='left')
```

```python
fig = px.scatter(players_accuracy,
                 y='TOTAL_POINTS',
                 x='ACCURACY %',
                 title='Distribution of points scored based on player accuracy',
                 labels={'ACCURACY %' : 'Accuracy', 'TOTAL_POINTS' : 'Sum of acquired points', 'FAVOURED_POSITION' : 'Position'},
                 color = 'FAVOURED_POSITION'
                 )

fig.add_vline(
    x= mean_accuracy,
    line_dash="dash",
    line_color="red",
    annotation_text=f"Mean accuracy {mean_accuracy:.2f}",
    annotation_position='bottom right'
    )

fig.show()
```

```python
home_analysis = df[df['TEAM_NAME_ABB'] == df['HOME_TEAM']]
home_analysis = home_analysis.groupby(['TEAM_NAME','EVENT_TYPE']).size().unstack(fill_value=0)


home_analysis['HOME_ACCURACY'] = (home_analysis['Made Shot'] / (home_analysis['Made Shot'] + home_analysis['Missed Shot']))
home_analysis = (home_analysis['HOME_ACCURACY'] * 100).round(2)
home_analysis = home_analysis.reset_index()


away_analysis = df[df['TEAM_NAME_ABB'] == df['AWAY_TEAM']]
away_analysis = away_analysis.groupby(['TEAM_NAME','EVENT_TYPE']).size().unstack(fill_value=0)


away_analysis['AWAY_ACCURACY'] = (away_analysis['Made Shot'] / (away_analysis['Made Shot'] + away_analysis['Missed Shot']))
away_analysis = (away_analysis['AWAY_ACCURACY'] * 100).round(2)
away_analysis = away_analysis.reset_index()


home_away_analysis = home_analysis[['TEAM_NAME', 'HOME_ACCURACY']].merge(away_analysis[['TEAM_NAME', 'AWAY_ACCURACY']], on= 'TEAM_NAME', how='left')

home_away_analysis['ACCURACY_DIF'] = home_away_analysis['HOME_ACCURACY'] - home_away_analysis['AWAY_ACCURACY']

home_away_analysis = home_away_analysis.sort_values(by='ACCURACY_DIF', ascending=False)

fig = px.bar(home_away_analysis,
             x='TEAM_NAME',
             y=['HOME_ACCURACY', 'AWAY_ACCURACY'],
             title='Team accuracy in 2023/2024 season based on home/away team infromation',
             labels={'value': 'Accuracy [%]', 'TEAM_NAME': 'Team name', 'HOME_ACCURACY':'Home Team Accuracy', 'AWAY_ACCURACY': 'Away Team Accuracy' }
             )

fig.update_layout(
    legend_title_text='Accuracy Type',
    barmode='group')

fig.add_vline(x='LA Clippers')

fig.show()
```

It would seem that in general being a Home Team helps team being more efficient.
More than half of the teams gets a boost in accuracy while playing on their own court.
